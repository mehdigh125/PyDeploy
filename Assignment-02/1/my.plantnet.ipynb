```python
import os
import requests
import dotenv
url = "https://my-api.plantnet.org/v2/identify/all"
PlantNet_api = os.getenv("PlantNet_api")
payload = {
    "api-key":PlantNet_api
}

headers = {
}

files = {
    'images': open('../image/lilium4.png','rb')
}
try:
    response = requests.post(url, params=payload, headers=headers, files=files)
    flowname = response.json()["results"][1]["species"]["genus"]["scientificName"]
    print(f"The name of the plant you sent is: {flowname}")
    
except:
    print(f"Dear user, your request has encountered error number {response.status_code},Please try again.")



```
